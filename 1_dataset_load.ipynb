### Pobranie i wstępna analiza danych

```python
import pandas as pd
import matplotlib.pyplot as plt # import umożliwia podpisanie osi wykresów

# Pobranie danych z pliku csv¶
df = pd.read_csv("dataset/languages.csv")
```

```python
# wyświetlenie 10 pierwszych wierszy z data frame
df.head(n=10)
```

```python
# wyświetlenie jednego z elementów z kolumny o nazwie text
df.text[5]
```

```python
# wyświetlenie histogramu z rokładem długości tekstu w kolumnie text
df.text.str.len().hist(bins = 100)
plt.xlabel('Ilość znaków w tekście')
plt.ylabel('Liczba wystąpień')
plt.title('Histogram długości tekstu')
```

```python
# ile języków występuje w df
len(df.labels.unique())
```

```python
# policzenie ilości wystąpień poszczególnych textów dla danego języka
df.labels.value_counts()
```

```python
len(df) #długość df - ile łącznie textów jest w df
```

### Przygotowanie i zapisanie przygotowanych danych

```python
# tutaj skracam ilości znaków w tekstach do maksymalnie 250
df['text'] = df['text'].str.slice(0, 250)
df.head(n=10)
```

```python
# wyświetlenie histogramu z rokładem długości tekstu w kolumnie text
df.text.str.len().hist(bins = 100)
plt.xlabel('Ilość znaków w tekście')
plt.ylabel('Liczba wystąpień')
plt.title('Histogram długości tekstu')
```

```python
# I wykres słupkowy
df.labels.value_counts().plot(kind = 'bar')
plt.ylabel('Liczba wystąpień')
plt.title('Wykres słupkowy ilości przykładów dla danego języka')
```

```python
# W ten sposób można spojrzeć na różne przykłady
df.text[df.labels == 'pl'].iloc[0]
```

```python
# Zapisuję df jako landuages_prep.csv - przygotowane do dalszej obróbki
df.to_csv("dataset/languages_prep.csv")
```
